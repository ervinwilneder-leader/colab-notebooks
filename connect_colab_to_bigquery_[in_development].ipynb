```python
#@title ## Autenticación
#@markdown ##### Antes de comenzar, ejecutar esta celda. Google Colab solicitará permisos de acceso. 
from google.colab import auth
auth.authenticate_user()

from google.cloud import bigquery
PROJECT_ID = 'leaderanalytics' #@param {'type': 'string'}
bigquery.magics.context.project = PROJECT_ID
client = bigquery.Client(project = PROJECT_ID)
```

## Ingesta de datos

Hay dos formas principales de acceder a los datos en BigQuery:

* **%%bigquery Magic**: Se utiliza para marcar el inicio de una sentencia SQL que se ejecutará tal como se haría en la UI de BigQuery. Admite una serie de parámetros según puede consultarse aquí https://googleapis.dev/python/bigquery/latest/magics.html#module-google.cloud.bigquery.magics.magics

* **Bigquery Python Client**: Es la forma "clásica". Se define inicialmente una sentencia SQL como string de Python y luego, mediante el método *query()*, la misma puede ejecutarse. Más detalles aquí https://cloud.google.com/bigquery/docs/visualize-jupyter#querying-and-visualizing-bigquery-data-using-pandas-dataframes

La primera de las opciones es más sencilla y práctica, pero limitada y menos versátil. A continuación, 2 ejemplos:

```python
# Se inicia con el magic y se usa el parámetro para definir un dataframe de salida, en este caso llamado "df" 
%%bigquery df 

SELECT * 
FROM youtube.vw_video_mappping 

# Pueden agregarse comentarios a lo largo de la celda, pero no código Python
```

```python
# El dataframe queda instanciado en la notebook y puede utilizarse como se lo haría habitualmente
df.head()
```

```python
# Similar al flujo anterior, usando el cliente de BigQuery para Python se puede hacer lo mismo
sql = """
SELECT * 
FROM youtube.vw_video_mappping 
"""

df = client.query(sql).to_dataframe()
df.head()
```
